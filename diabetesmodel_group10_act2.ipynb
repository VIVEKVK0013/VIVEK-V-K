```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("diabetes_model.csv")
```

```python
data
```

# 1.Create a Count plot of Outcome

```python
sns.countplot(data["Outcome"])
```

# 2.Analyse the effect of age on Outcome using Boxplot

```python
sns.boxplot(x="Outcome",y="Age",data=data)
```

# 3.Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function

```python
sns.swarmplot(x="Outcome",y="DiabetesPedigreeFunction",data=data)
```

# 4.Check for the reletionship with Insulin and Glucose and group the datapoints by Outcome

```python
sns.scatterplot(x=data["Insulin"],y=data["Glucose"],hue=data["Outcome"],data=data)
```

# 5.Create a heatmap on Correlation of features in data

```python
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=90)
```
